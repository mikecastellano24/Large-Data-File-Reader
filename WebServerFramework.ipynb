# Python

```python
from flask import Flask, request, render_template
from google_drive_handler import handle_google_drive_link
# Add other libraries from PDF Regex Reader

app = Flask(__name__)

@app.route('/', methods=['GET', 'POST'])
def submit_drive_link():
    if request.method == 'POST':
        url = request.form['url']
        keyword = request.form['keyword']

        ###
        # PDF KEYWORD READER HERE
        ###

        results = handle_google_drive_link(url, keyword)
        return str(results)  # return results as a string
    return render_template('submit_link.html')  # Render the form
```
